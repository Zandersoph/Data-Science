```python
import pandas
```

```python
from pandas import DataFrame
import matplotlib.pyplot as mpy
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_dirty.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=["production_budget_usd"])
Y = DataFrame(data, columns=["worldwide_gross_usd"])
```

```python
mpy.figure(figsize=(10,6))
mpy.scatter(X,Y,alpha=0.25)
mpy.title("Film Costs VS Global Revenue")
mpy.xlabel("Production Budget $")
mpy.ylabel("Global Revenue $")
mpy.ylim(0,3000000000)
mpy.xlim(0,450000000)
mpy.show()
```

```python
regression = LinearRegression()
```

```python
regression.fit(X,Y)
```

Slope Coefficient:
h(X) = theta0 + theta1.X

```python
regression.coef_ #theta1
```

```python
regression.intercept_ #theta0
```

```python
mpy.figure(figsize=(10,6))
mpy.plot(X,regression.predict(X), color = 'red', linewidth = 4)
mpy.scatter(X,Y,alpha=0.25)
mpy.title("Film Costs VS Global Revenue")
mpy.xlabel("Production Budget $")
mpy.ylabel("Global Revenue $")
mpy.ylim(0,3000000000)
mpy.xlim(0,450000000)
mpy.show()
```

```python
regression.score(X,Y)
```
