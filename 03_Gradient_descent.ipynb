## Gradient descent as a python function

```python
#Gradient Descent function
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter = 300):

    new_x = initial_guess
    step_multiplier = multiplier

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - step_multiplier*gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=0.5, multiplier=0.02, precision=0.001)
print("Local minima occurs at: ", local_min)
print("Number of steps: ", len(list_x))
```

## Minimise $$f(x,y) = \frac{1}{3^{-x^2-y^2}+1}$$

Minimise $$f(x,y) = \frac{1}{r+1}$$ where ${r}$ is ${3^{-x^2-y^2}}$

```python
def f(x,y):
    r = 3**(-x**2-y**2)
    return 1/(r+1)
```

```python
# Make x and Y data
x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)

# print("Shape of X Array",x_4.shape)

x_4, y_4 = np.meshgrid(x_4, y_4)
print("Shape of X Array",x_4.shape)
```

## Partial derivative and symbolic computations
## $$\frac{\partial f}{ \partial x} = \frac{2x \cdot \ln(3) \cdot 3^{-x^2 -y^2}}{\left(3 ^ {-x^2-y^2} +1 \right)^2}$$
## $$\frac{\partial f}{ \partial y} = \frac{2y \cdot \ln(3) \cdot 3^{-x^2 -y^2}}{\left(3 ^ {-x^2-y^2} +1 \right)^2}$$

# Batch gradient descent with SymPy

```python
a, b = symbols('x,y')
x_pdev = diff(f(a,b),a)
y_pdev = diff(f(a,b),b)
print('Our cost function is: ',f(a,b))
print('Partial derivative w.r.t x is: ',x_pdev)
print('Partial derivative w.r.t y is: ',y_pdev)
print("Value of f(x,y) at x=1.8 and y=1.0 is: ", f(a,b).evalf(subs={a:1.8,b:1.0}))
print("Slope of x at this point is: ",x_pdev.evalf(subs={a:1.8,b:1.0}))
```

```python
#Setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = x_pdev.evalf(subs={a:params[0],b:params[1]})
    gradient_y = y_pdev.evalf(subs={a:params[0],b:params[1]})
    gradients = np.array([gradient_x, gradient_y])
    
    #Learning step
    params = params-multiplier*gradients

#Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

```python
#Partial derivative functions
def fpx(x,y):
    num = 2*x*log(3)*3**(-x**2-y**2)
    denom = (3**(-x**2-y**2)+1)**2
    return num/denom

def fpy(x,y):
    num = 2*y*log(3)*3**(-x**2-y**2)
    denom = (3**(-x**2-y**2)+1)**2
    return num/denom
```

```python
#Setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    
    #Learning step
    params = params-multiplier*gradients

#Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

## MSE: A cost function for regression problem

### $$RSS = \sum_{i=1}^{n} \big(y^{(i)} - h_\theta x^{(i)} \big)^2$$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big(y^{(i)} - h_\theta x^{(i)} \big)^2$$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big(y - \hat{y} \big)^2$$

```python
# Make sample data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([[1.7,2.4,3.5,3.0,6.1,9.4,8.2]]).reshape(7,1)
print('Shape of x_5 array:', x_5.shape)
print('Shape of y_5 array:', y_5.shape)
```

```python
#Linear regression
regr = LinearRegression()
regr.fit(x_5, y_5)
print('Theta0: ', regr.intercept_[0])
print('Theta1: ', regr.coef_[0][0])
```

```python
plt.scatter(x_5,y_5, s=50)
plt.plot(x_5, regr.predict(x_5), color='orange', linewidth=3)
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.show()
```

```python
#y_cap = theta0 + theta1*x
y_cap = 0.8475351486029536 + 1.2227264637835915*x_5
print('Estimated values of y_hat are: \n', y_cap)
print('In comparison, the actual y_values are \n', y_5)
```

```python
def MSE(y, y_cap):
    res = 0
    for i in range(len(y_cap)):
        res += (y[i]-y_cap[i])**2
    return res/len(y_cap)
```

```python
print('Manually calculated MSE is: ', MSE(y_5, y_cap))
print('MSE Regression using manual calc is', mean_squared_error(y_5, y_cap))
print('MSE Regresion is: ', mean_squared_error(y_5, regr.predict(x_5)))
```

## Partial Derivative of MSE w.r.t $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = -\frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = -\frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \cdot \big(x^{(i)} \big)$$

## MSE and Gradient descent

```python
#x-values, y-values, array of theta parameters(theta0 - index[0] and theta1 - index[1])
def grad(x,y,thetas):
    n = y.size
    theta0_slope = (-2/n)*sum(y-thetas[0]-thetas[1]*x)
    theta1_slope = (-2/n)*sum((y-thetas[0]-thetas[1]*x)*x)
    
    #return np.array([theta0_slope[0], theta1_slope[o]])
    return np.append(arr=theta0_slope, values=theta1_slope)
    #return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

#Collect datapoints for scatter
plot_vals = thetas.reshape(1,2)
mse_vals = MSE(y_5, thetas[0]+thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier*grad(x_5, y_5, thetas)
    #append new values to array
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1,2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=MSE(y_5, thetas[0]+thetas[1]*x_5))

print(f'Min occurs ar Theta 0: {thetas[0]}')
print(f'Min occurs at Theta 1: {thetas[1]}')
print(f'MSE is: {MSE(y_5, thetas[0]+thetas[1]*x_5)}')
```
